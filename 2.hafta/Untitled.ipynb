```python
#gönderilen mxn lik resmi m/2 x n/2 lik yapan fonksiyon

def boyutlandir(img_4):
    m,n,p=img_4.shape
    new_m=int(m/2)
    new_n=int(n/2)
    img_5=np.zeros((new_m,new_n),dtype=float)
    for m in range(new_m): 
        for n in range(new_n):
            s=float((img_4[m*2,n*2,0]+img_4[m*2,n*2,1]+img_4[m*2,n*2,2])/3)
            img_5[m,n]=float(s)
    return img_5
```

```python
img_6=boyutlandir(img_1)
plt.imshow(img_6,cmap='gray')
plt.show()
```

```python
# YENİ RESİM OKUDUK


img_yeni=plt.imread('images.jpg')
```

```python
def bozulmadan_boyutlandir(im_7):
    m,n,p=im_7.shape
    new_m=int(m/2)
    new_n=int(n/2)
    im_8=np.zeros((new_m,new_n,3),dtype=int)
    for m in range(new_m): 
        for n in range(new_n):
            im_8[m,n,0]=int(im_7[m*2,n*2,0])
            im_8[m,n,1]=int(im_7[m*2,n*2,1])
            im_8[m,n,2]=int(im_7[m*2,n*2,2])
    return im_8

```

```python
plt.imshow(img_yeni)
plt.show()
```

```python
img_boyutlandirilmis=bozulmadan_boyutlandir(img_yeni)
plt.imshow(img_boyutlandirilmis)
plt.show()
```

```python
img_boyutlandirilmis_2=bozulmadan_boyutlandir(img_boyutlandirilmis)
plt.imshow(img_boyutlandirilmis_2)
plt.show()
```
