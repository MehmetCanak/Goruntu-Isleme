```python
import numpy as np
import matplotlib.pyplot as plt
img=plt.imread('kis.png')
type(img)
img.shape
```

```python
img[100,100,:]
```

```python
m,n,p = img.shape
m,n,p
```

```python
# bir resmi siyah beyaz yapmak

new_img=np.zeros((m,n),dtype=float) 

for i in range (m):
    for j in range (n):
        s=(img[i,j,0]+img[i,j,1]+img[i,j,2])/3
        new_img[i,j]=s

plt.imshow(new_img,cmap='gray')
plt.show() 
plt.imsave('siyahbeyaz.png',new_img,cmap='gray')
```
